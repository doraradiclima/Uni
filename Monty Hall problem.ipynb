Kao sto i Monte Carlo simulacija pokazuje vjeroatnost osvajanja auta u slucaj da ne promjenimo vrata je 1/3 dok je vjeroatnost pobjede auta 2/3 u slucaju da zamjenimo vrata. Kada prvi put biramo vrata imamo 1/3 vjeroatnosti da osvojimo dok druga dva vrata imaj vjeroatnost od 2/3 da je iza njih auto. Nakon sto Monty Hall eliminira jedna od ta vrata vjeroatnost da je iza preostalih vrata auto je i dalje 2/3.   

```python
import numpy as np
from random import randint
def nepo(p,n):
    return np.sqrt((p*(1-p))/(n-1))
    
zamjena=0
isto=0
n=1000
for i in range(1000):
    vrata = np.array([1,0,0])
    np.random.shuffle(vrata)
    prvi_pogodaj = randint(0, 2)

    for k in range(0,3):
        if vrata[k]!=1 and k!=prvi_pogodaj:
            otvorena=k
            break

    for k in range(0,3):
        if vrata[k]==1:
            auto=k

    if auto==prvi_pogodaj:
        isto+=1

    if auto==3-otvorena-prvi_pogodaj:
        zamjena+=1

print("Ako promjeni vrata ima vjv od %.4f %% za pobjedu uz nepouzdanost od %.4f %%" %(zamjena/n*100, nepo(zamjena/n,n)*100))
print("Ako nepromjeni vrata ima vjv od %.4f %% za pobjedu uz nepouzdanost od %.4f %%" %(isto/n*100, nepo(isto/n,n)*100))
#print(isto/n,nepo(isto/n,n), zamjena/n, nepo(zamjena/n,n))
```
