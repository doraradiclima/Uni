```python
import numpy as np
import itertools
import matplotlib.pyplot as plt

N=7
j=1

#svi mogući rastavi broja N na zbroj prirodnih brojeva

def rastavi_broja_N(N, i = 1):  
   
    if (N == 0):
        if (len(pribrojnici) != 1):
            permutacije()
                  
    for j in range(i, N + 1):
        pribrojnici.append(j)
        rastavi_broja_N(N-j, j)  
        
        del pribrojnici[len(pribrojnici)-1]
    
    
def permutacije():
    permutacije_pribrojnika = itertools.permutations(pribrojnici)
    jedan_rastav = list(set(permutacije_pribrojnika)) #jedan rastav na brojeve, bez duplikata
    matrica.append(jedan_rastav)
    
    return matrica
   
pribrojnici = []
matrica=[]
rastavi = [] #matrica svih mogućih rastava broja N na zbroj prirodnih brojeva
    
rastavi_broja_N(N)


if(N==1): 
    rastavi=[1] #trivijalno
    print(lista)
else:    
    for i in range(len(matrica)):
        rastavi += matrica[i]
    M=[N]    
    rastavi.append(M)    

for i in range(len(rastavi)):
     rastavi[i]=list(rastavi[i])
    
print("Svi mogući rastavi broja N na zbroj prirodnih brojeva:\n",rastavi)
```

Vjerojatnost prijelaza i → j iz nekog stanja i u neko niže stanje j je pij. Zanima nas proučiti kad su vjerojatnosti bilo koje cjelovite deekscitacije u osnovno stanje jednake.

Ako prvo promotrimo slučaj kada je N=1: jedina moguća cjelovita deekscitacija je 1 → 0. Vjerojatnost tog prijelaza je onda 1. Uočavamo da, od kojeg god N krenuli, ako nakon m skokva dođemo do 1 → 0, vjerojatnost tog prijelaza p(1→0) je 1. 

Promotrimo sada slučaj N=2: moguće cjelovite deekscitacije su 2 → 1 → 0 i 2 → 0. Znamo da ove dvije deekscitacije moraju biti jednako vjerojatne iz čega slijedi prva jednakost (1). Nadalje znamo da je vjerojatnost prijelaza 1 → 0 jednaka 1. Kao nepoznanice nam ostaju vjerojatnosti prijelaza p(2→1) i p(2→0). Za cjelovitu deekscitaciju 2 → 1 → 0, vjerojatnosti p(2→1) i p(1→0) međusobno množimo. Konačno iz uvjeta da zbroj vjerojatnosti svih pojedinih deekscitacija mora biti 1 slijedi druga jednakost (2).

     
(1): p(2→1→0)=p(2→0) <br>
(2): p(2→1→0)+p(2→0)=1 <br><br>

$(2→1) * p(1→0) = p(2→0)$ <br>
$p(2→1) * p(1→0) + p(2→0) = 1$<br><br>
$p(2→1) * 1 = p(2→0)$<br>
$p(2→1) * 1 + p(2→0) = 1$<br><br>
uvrštavanjem (1) u (2) imamo: <br>
$p(2→0) + p(2→0) = 1$<br> 
$2p(2→0) = 1  /:2$. <br>
$p(2→0) = 0.5$ <br><br>
iz čega slijedi da je i:  <br>
$p(2→1)=0.5$ <br>

Na temelju ovoga možemo formirati matricu prijelaza tako da broj retka predstavlja i, a broj stupca j. <br>
$ \begin{bmatrix}
0 & 0 & 0 \\
1 & 0 & 0 \\
0.5 & 0.5 & 0 
\end{bmatrix}  $
<br><br>
Za veće N matrica se slaže po uočenom pravilu.
                                                                                                          

```python
def vjerojatnosti_svih_prijelaza():
    
    matrica_p = np.zeros((N+1,N+1))

    for j in range(N+1):
        for i in range(N+1):
            if j==0 and i>0:
                matrica_p[i][j]=1/(2**(i-1))
            elif j==i-1:
                matrica_p[i][j]=0.5
            else:
                matrica_p[i][j]=matrica_p[i-1][j]/2
   
    return matrica_p       

matrica=vjerojatnosti_svih_prijelaza()

print('Matrica vjerojatnosti svih prijelaza p_ij za  N =',N,':\n')
print(matrica, '\n')

for i in range (N+1):
    for j in range(N+1):
        if i>j:
            print('p (',i,'→',j,'):',matrica[i][j])
```

```python
print(matrica)
indeksi_za_matricu = []
vjerojatnosti_iz_matrice = []

for i in range(len(rastavi)):
    n = N
    pomocno = []
    pomocno.append(n)
    for j in rastavi[i]:
        n=n-j
        pomocno.append(n)
    indeksi_za_matricu.append(pomocno)
print(indeksi_za_matricu)

for i in range(len(rastavi)):
    pomocno = []
    for j in range(len(rastavi[i])):
        pomocno.append(matrica[indeksi_za_matricu[i][j]][indeksi_za_matricu[i][j+1]])
    vjerojatnosti_iz_matrice.append(pomocno)
        
print(vjerojatnosti_iz_matrice)
```

```python
def prosjecan_broj_fotona_po_kaskadi():
    broj_fotona=0
    broj_kaskada=len(rastavi)
    
    for i in range(len(rastavi)):
        broj_fotona+=len(rastavi[i])
   
    return broj_fotona/broj_kaskada

def prosjecna_energija_fotona():
    E=[4,6,11,15,22,34,50,66,72,100, 234]
    E=E[:N+1]
    
    vjerojatnost_slucaj_1=[]
    vjerojatnost_slucaj_2=vjerojatnosti_iz_matrice
    
    for i in range(len(rastavi)):
        pom=[]
        for j in range(len(rastavi[i])):
            pom.append(1 / indeksi_za_matricu[i][j])
        vjerojatnost_slucaj_1.append(pom)
    
    razlika_skokova=[]
    
    for i in range(len(indeksi_za_matricu)):
        pomocno=[]
        for j in range(len(indeksi_za_matricu[i])-1):
            pomocno.append(E[indeksi_za_matricu[i][j]]-E[indeksi_za_matricu[i][j+1]])
        razlika_skokova.append(pomocno)

    
    prosjecne_energije_1 = []
    
    for i in range(len(razlika_skokova)):
        pomocno = 0
        for j in range(len(razlika_skokova[i])):
            vjer_uk = 1
            for k in range(j+1):
                vjer_uk *= vjerojatnost_slucaj_1[i][k]
            pomocno += (razlika_skokova[i][j] * vjer_uk)
        prosjecne_energije_1.append(pomocno)
         
        
    prosjecne_energije_2 = []
    
    for i in range(len(razlika_skokova)):
        pomocno = 0
        for j in range(len(razlika_skokova[i])):
            vjer_uk = 1
            for k in range(j+1):
                vjer_uk *= vjerojatnost_slucaj_2[i][k]
            pomocno += (razlika_skokova[i][j] * vjer_uk)
        prosjecne_energije_2.append(pomocno)
    
    
    return prosjecne_energije_1, prosjecne_energije_2

rezultat = prosjecna_energija_fotona()
plt.scatter(range(len(rezultat[0])), rezultat[0], color='r')
plt.scatter(range(len(rezultat[1])), rezultat[1], color='forestgreen')

prosjecna_energija_1=np.sum(rezultat[0])
prosjecna_energija_2=np.sum(rezultat[1])
    
print('prosjecan broj fotona po kaskadi: ',prosjecan_broj_fotona_po_kaskadi())  
print('prosjecna energija u slucaju 1:',prosjecna_energija_1,'\nprosjecna energija u slucaju 2: ', prosjecna_energija_2)
```
